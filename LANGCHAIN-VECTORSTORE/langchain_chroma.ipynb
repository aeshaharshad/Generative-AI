```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document

# Create documents
doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history.",
    metadata={"team": "Royal Challengers Bangalore"}
)

doc2 = Document(
    page_content="Rohit Sharma is the most successful captain in IPL history.",
    metadata={"team": "Mumbai Indians"}
)

doc3 = Document(
    page_content="MS Dhoni, famously known as Captain Cool.",
    metadata={"team": "Chennai Super Kings"}
)

docs = [doc1, doc2, doc3]

# Initialize FREE local embeddings
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Create vector store
vector_store = Chroma(
    embedding_function=embedding_model,
    persist_directory="my_chroma_db",
    collection_name="sample"
)

# Add documents
vector_store.add_documents(docs)

print("✅ Documents added successfully using FREE local embeddings!")
```

```python
results = vector_store.similarity_search("Who is captain of Mumbai?", k=2)

for r in results:
    print(r.page_content)
```

```python
# view documents
vector_store.get(include=['embeddings','documents', 'metadatas'])
```

```python
# search documents
vector_store.similarity_search(
    query='Who among these are a bowler?',
    k=2
)
```

```python
# search with similarity score
vector_store.similarity_search_with_score(
    query='Who among these are a bowler?',
    k=2
)
```

```python
# meta-data filtering
vector_store.similarity_search_with_score(
    query="",
    filter={"team": "Chennai Super Kings"}
)
```

```python
# update documents
updated_doc1 = Document(
    page_content="Virat Kohli, the former captain of Royal Challengers Bangalore (RCB), is renowned for his aggressive leadership and consistent batting performances. He holds the record for the most runs in IPL history, including multiple centuries in a single season. Despite RCB not winning an IPL title under his captaincy, Kohli's passion and fitness set a benchmark for the league. His ability to chase targets and anchor innings has made him one of the most dependable players in T20 cricket.",
    metadata={"team": "Royal Challengers Bangalore"}
)

vector_store.update_document(document_id='df5dbf4f-9276-45b9-a9f1-c4cb93f0f617', document=updated_doc1)
```

```python
#view documents
vector_store.get(include=['embeddings','documents', 'metadatas'])
```
