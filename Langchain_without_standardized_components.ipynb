#make LLM component- not real which can be used by ai engineer to interact with llm , connect components manually by code


```python
import random
class DummyLLM:
  def __init__(self):
    print("LLM created")
  def predict(self,prompt): # as llm component has predict method
    response_list=[
        'Ankara is capital of Turkey',
        'Research internships are valueable',
        'PSL stands for Pakistan Super League'
    ]
    return {'response':random.choice(response_list)}
```

```python
llm=DummyLLM()
```

```python
llm.predict('What is famous Landmark in spain')
```

```python
class DummyPromptTemplate:
  def __init__(self,template,input_variables):
    self.template=template
    self.input_variables=input_variables
  def format(self,input_dict):
    return self.template.format(**input_dict)
```

```python
template=DummyPromptTemplate(
    template='What is the capital of {country}',
    input_variables=['country']
)
```

```python
prompt=template.format({'country':'Belgium'})
```

```python
llm=DummyLLM()
```

```python
llm.predict(prompt)
```

Dummy LLM chain-not flexible to create workflows , since its chain it has run method


```python
class DummyLLMChain:
  def __init__(self,llm,prompt):
    self.llm=llm
    self.prompt=prompt
  def run(self,input_dict):
    #form prompt from our prompt template
    final_prompt=self.prompt.format(input_dict)
    result=self.llm.predict(final_prompt) # result from llm is in dictionary format,we extract string
    return result['response']
```

```python
template=DummyPromptTemplate(
    template='What is the capital of {country}',
    input_variables=['country']
)
```

```python
llm=DummyLLM()
```

```python
chain=DummyLLMChain(llm,template)
```

```python
chain.run({'country':'Pakistan'})
```
